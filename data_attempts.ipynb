```python
def flatten_dict(d: dict) -> dict:
    out = {}
    for k, v in d.items():
        if isinstance(v, dict):
            for k2, v2 in flatten_dict(v).items():
                out[f'{k}.{k2}'] = v2
        else:
            out[k] = v
    return out
```
